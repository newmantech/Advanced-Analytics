```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
path = r'C:\Users\dbnew\Achievement_6'
```

```python
#Importing WHO air quality dataset
df = pd.read_excel(os.path.join(path, 'Original_Data', 'who_orig.xlsx'), index_col = False)
```

```python
df.head()
```

```python
#Dropping unneeded columns
df=df.drop(columns=['ISO3', 'Status'])
```

```python
df.head()
```

```python
df.describe()
```

```python
#Check for mixed data types
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
#Changing WHO Region column to string type
df['WHO Region'] = df['WHO Region'].astype('str')
```

```python
df['Number and type of monitoring stations'].dtype
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
#Checking for pattern in missing PM2.5 values
df_nan_pm2=df[df['PM2.5 (μg/m3)'].isnull()==True]
```

```python
#Counting missing PM2.5 values by country
missing_data_by_country=df_nan_pm2['WHO Country Name'].value_counts(dropna=False).to_frame()
```

```python
pd.set_option('display.max_rows', None)
```

```python
missing_data_by_country
```

```python
#Checking total entries for each country to establish percentage of missing values
df_full=df['WHO Country Name'].value_counts(dropna=False).to_frame()
```

```python
df_full
```

```python
#Caclulating percentage missing PM2.5 values for each country
df_missing_perc=missing_data_by_country/df_full
```

```python
df_missing_perc
```

```python
#Sorting percentages of missing PM2.5 values in descending order
sorted_df = df_missing_perc.sort_values(by='WHO Country Name', ascending=False)
```

```python
sorted_df
```

We will consider removing the top 14 countries listed above for any computations involving PM2.5 measurements because more than 95% of the data is missing for these countries.
